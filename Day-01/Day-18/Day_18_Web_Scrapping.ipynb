```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get("https://www.scrapethissite.com/pages/").text
```

```python
soup = BeautifulSoup(webpage , 'lxml')
```

```python
scrapping = soup.find_all('div', class_='page')
```

```python
len(scrapping)
```

```python
title = []
content = []
for i in scrapping:
    title.append(i.find('h3').text.strip())
    content.append(i.find('p').text.strip())
```

```python
d = {'title':title , 'content':content}
```

```python
df = pd.DataFrame(d)
```
